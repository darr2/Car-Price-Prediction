```python
#function definition which takes a pandas series and a datatype as input
#outputs a tuple with number of values present, number of values not present
#and percentage of number of values not present
def missing_value_pct(series, val_type):
    has_value, has_no_value = 0, 0
    for value in series:
        if type(value) == val_type:
            has_value+=1
        elif np.isnan(value):
            has_no_value+=1
        else:
            has_no_value+=1
    return (has_value, has_no_value, (has_no_value/(has_value+has_no_value)))


#Check for missing values     
print('Make')
print(missing_value_pct(car_df_2017['Make'], str))
print('Model')
print(missing_value_pct(car_df_2017['Model'], str))
print('Year')
print(missing_value_pct(car_df_2017['Year'], int))
print('Engine Fuel Type')
print(missing_value_pct(car_df_2017['Engine Fuel Type'], str))
print('Engine HP')
print(missing_value_pct(car_df_2017['Engine HP'], float))
print('Engine Cylinders')
print(missing_value_pct(car_df_2017['Engine Cylinders'], float))
print('Transmission Type')
print(missing_value_pct(car_df_2017['Transmission Type'], str))
print('Driven_Wheels')
print(missing_value_pct(car_df_2017['Driven_Wheels'], str))
print('Number of Doors')
print(missing_value_pct(car_df_2017['Number of Doors'], float))
print('Market Category')
print(missing_value_pct(car_df_2017['Market Category'], str))
print('Vehicle Size')
print(missing_value_pct(car_df_2017['Vehicle Size'], str))
print('Vehicle Style')
print(missing_value_pct(car_df_2017['Vehicle Style'], str))
print('highway MPG')
print(missing_value_pct(car_df_2017['highway MPG'], int))
print('city mpg')
print(missing_value_pct(car_df_2017['city mpg'], int))
print('Popularity')
print(missing_value_pct(car_df_2017['Popularity'], int))
print('MSRP')
print(missing_value_pct(car_df_2017['MSRP'], int))
```

```python
#Exposition, summary, explanation of data and relationship
#Make explanation to be two sentences
#This is a plot of all the avaialbe car brands in the dataset.
#Their respective pricing distributions are shown as violin plots such that the range, spread, and
#median values are compactly represented.  As we can see here, luxury brand cars such as Mercedes, Maserati,
#and Porsche have a larger range of data and higher MSRP medians than their compact competitors.
plt.xlabel('Make', fontsize=40)
plt.ylabel('Average MSRP', fontsize=40)
plt.xticks(fontsize=30,rotation=90)
plt.yticks(fontsize=20,rotation=0)
sns.violinplot('Make', 'MSRP', order=sorted(list(car_df_2017.Make.unique())), data=car_df_2017)
sns.despine(trim=True)
plt.rcParams["figure.figsize"] = (30,10)
plt.show()
```

```python
#Plot average price vs Make and Transmission
#Shown here are violin plots that visualize the distribution of MSRP with respect to transmission.
#What can be observed here is that manual transmissions tend to have a higher probability of being
#cheaper than automatic, direct drive, and automated manual transmissions.  
plt.xlabel('Transmission Type', fontsize=20)
plt.ylabel('Average Price', fontsize=20)
sns.violinplot('Transmission Type', 'MSRP', data=car_df_2017)
sns.despine(trim=True)
plt.rcParams["figure.figsize"] = (10, 5)
plt.show()
```

```python
#The following two plots show city MSRP vs city mpg and MSRP vs highway mpg, respectively.
#This is plotted using a Kernel Density Estimation plots.  What we notice here is that
#vehicles with with higher MPGs tend to have lower MSRPs compared to vehicles with higher 
#MPGs.  This comes from the fact that luxury and performance cars tend to be less fuel
#efficient and usually more expensive than compact economy cars.
sns.jointplot(x='city mpg', y='MSRP', data=car_df_2017)
sns.jointplot(x='highway MPG', y='MSRP', data=car_df_2017)
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore')
```

```python
#Here we have another set of violin plots showing the MSRP distribution
#with respect to the number of doors for any given vehicle.  There is ma much
#larger spread in MSRP for two door vehicles than four door vehicles and the observed
#median and mean for two door vehicles is slightly higher.  One very intersting
#observation is that there are three door vechiles in the data set!
sns.violinplot('Number of Doors', 'MSRP', data=car_df_2017)
sns.despine(trim=True)
plt.rcParams["figure.figsize"] = (5, 5)
plt.show()
```
